```python
from pathlib import Path
import sys
import importlib

root = Path.cwd()
if (root / 'cases').exists():
    project_root = root
else:
    project_root = root.parent
sys.path.append(str(project_root))

case_dir = project_root / 'cases' / 'supply-chain-analysis'
data_path = case_dir / 'supply_chain_data.csv'
module_base = 'cases.supply-chain-analysis.src'
```

```python
data_loading = importlib.import_module(f"{module_base}.data_loading")
cleaning = importlib.import_module(f"{module_base}.cleaning")
features = importlib.import_module(f"{module_base}.features")
modeling = importlib.import_module(f"{module_base}.modeling")

df = features.add_features(cleaning.clean_data(data_loading.load_raw(data_path)))
scores_df, importance_df, metrics = modeling.run_driver_model(df, target='revenue_generated')
metrics
```
