```
from pathlib import Path
import sys
import importlib

root = Path.cwd()
if (root / 'cases').exists():
    project_root = root
else:
    project_root = root.parent
sys.path.append(str(project_root))

case_dir = project_root / 'cases' / 'procurement-kpi-analysis'
data_path = case_dir / 'Procurement KPI Analysis Dataset.csv'
module_base = 'cases.procurement-kpi-analysis.src'
```

```
data_loading = importlib.import_module(f"{module_base}.data_loading")
cleaning = importlib.import_module(f"{module_base}.cleaning")
features = importlib.import_module(f"{module_base}.features")
modeling = importlib.import_module(f"{module_base}.modeling")

raw = data_loading.load_raw(data_path)
df = features.add_features(cleaning.clean_data(raw))
```

```
supplier_df = modeling.supplier_metrics(df)
supplier_df = modeling.supplier_risk_score(supplier_df)
supplier_df = modeling.segment_suppliers(supplier_df)

supplier_df.head()
```

```
pareto_savings = modeling.pareto_by_metric(supplier_df, 'realized_savings')
pareto_savings.head()
```

```
scenario1 = modeling.scenario_noncompliant_spend(supplier_df)
scenario2 = modeling.scenario_defect_reduction(supplier_df)
scenario1, scenario2
```
